## 3. Imports et initialisation


```python
import cv2
import dlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
```

## 4. Détection des landmarks

```python
def get_landmarks(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    rects = detector(gray, 1)
    if len(rects) == 0: return None
    shape = predictor(gray, rects[0])
    coords = np.zeros((68, 2), dtype="int")
    for i in range(68):
        coords[i] = (shape.part(i).x, shape.part(i).y)
    h, w, _ = image.shape
    corners = np.array([[0,0],[w-1,0],[0,h-1],[w-1,h-1],
                         [w//2,0],[w//2,h-1],[0,h//2],[w-1,h//2]])
    return np.concatenate((coords, corners), axis=0)
```

## 5. Transformation affine

```python
def apply_affine_transform(src, src_tri, dst_tri, size):
    if size[0] <= 0 or size[1] <= 0:
        return np.zeros((max(size[1],1), max(size[0],1), 3), dtype=np.float32)
    if src.shape[0] <= 0 or src.shape[1] <= 0:
        return np.zeros((size[1], size[0], 3), dtype=np.float32)
    warp_mat = cv2.getAffineTransform(np.float32(src_tri), np.float32(dst_tri))
    return cv2.warpAffine(src, warp_mat, (size[0], size[1]), None,
                          flags=cv2.INTER_LINEAR,
                          borderMode=cv2.BORDER_REFLECT_101)
```

## 6. Morphing par triangle

Le morphing se fait **triangle par triangle**. Pour chaque triangle :
1. On extrait la région correspondante dans chaque image source
2. On applique une transformation affine pour aligner les triangles
3. On fusionne les deux régions avec le paramètre `alpha`
4. On applique un masque pour ne modifier que la zone du triangle

```python
def morph_triangle(img1, img2, img, t1, t2, t, alpha):
    r  = cv2.boundingRect(np.float32([t]))
    r1 = cv2.boundingRect(np.float32([t1]))
    r2 = cv2.boundingRect(np.float32([t2]))


    if r[2] <= 0 or r[3] <= 0 or r1[2] <= 0 or r1[3] <= 0 or r2[2] <= 0 or r2[3] <= 0:
        return

    t_rect, t1_rect, t2_rect = [], [], []
    for i in range(3):
        t_rect.append(((t[i][0]  - r[0]),  (t[i][1]  - r[1])))
        t1_rect.append(((t1[i][0] - r1[0]), (t1[i][1] - r1[1])))
        t2_rect.append(((t2[i][0] - r2[0]), (t2[i][1] - r2[1])))

    mask = np.zeros((r[3], r[2], 3), dtype=np.float32)
    cv2.fillConvexPoly(mask, np.int32(t_rect), (1.0, 1.0, 1.0), 16, 0)

    img1_rect = img1[r1[1]:r1[1]+r1[3], r1[0]:r1[0]+r1[2]]
    img2_rect = img2[r2[1]:r2[1]+r2[3], r2[0]:r2[0]+r2[2]]


    if img1_rect.size == 0 or img2_rect.size == 0:
        return

    warp_img1 = apply_affine_transform(img1_rect, t1_rect, t_rect, (r[2], r[3]))
    warp_img2 = apply_affine_transform(img2_rect, t2_rect, t_rect, (r[2], r[3]))

    img_rect = (1.0 - alpha) * warp_img1 + alpha * warp_img2
    img[r[1]:r[1]+r[3], r[0]:r[0]+r[2]] = \
        img[r[1]:r[1]+r[3], r[0]:r[0]+r[2]] * (1 - mask) + img_rect * mask
```

## 7. Fonction principale de morphing

La fonction `face_morph` orchestre l'ensemble du processus :

1. Détection des landmarks sur les deux images
2. Calcul des landmarks interpolés `pm = (1-alpha)*p1 + alpha*p2`
3. **Triangulation de Delaunay** sur `pm` : on découpe l'espace en triangles optimaux
4. Application du morphing triangle par triangle
5. Correction des éventuelles zones noires par un blend direct

```python
def face_morph(img1, img2, alpha=0.5):
    p1 = get_landmarks(img1)
    p2 = get_landmarks(img2)
    if p1 is None or p2 is None:
        print(" Landmarks non détectés sur une des images.")
        return None

    h, w, _ = img1.shape
    pm = (1 - alpha) * p1 + alpha * p2

    pm[:, 0] = np.clip(pm[:, 0], 0, w - 1)
    pm[:, 1] = np.clip(pm[:, 1], 0, h - 1)
    p1[:, 0] = np.clip(p1[:, 0], 0, w - 1)
    p1[:, 1] = np.clip(p1[:, 1], 0, h - 1)
    p2[:, 0] = np.clip(p2[:, 0], 0, w - 1)
    p2[:, 1] = np.clip(p2[:, 1], 0, h - 1)

    subdiv = cv2.Subdiv2D((0, 0, w, h))
    for p in pm:
        subdiv.insert((int(p[0]), int(p[1])))

    tri_list = subdiv.getTriangleList()
    img_m = np.zeros(img1.shape, dtype=np.float32)

    for t in tri_list:
        pt = [(t[0], t[1]), (t[2], t[3]), (t[4], t[5])]
        ind = []
        for p in pt:
            dist = np.sum((pm - p)**2, axis=1)
            ind.append(np.argmin(dist))
        morph_triangle(
            np.float32(img1), np.float32(img2), img_m,
            [p1[i] for i in ind],
            [p2[i] for i in ind],
            [pm[i] for i in ind],
            alpha
        )

    img_m_uint8 = np.uint8(img_m)
    mask_black = np.all(img_m_uint8 < 10, axis=2)
    blend_direct = np.uint8((1 - alpha) * np.float32(img1) + alpha * np.float32(img2))
    img_m_uint8[mask_black] = blend_direct[mask_black]
    img_m_uint8 = cv2.GaussianBlur(img_m_uint8, (3, 3), 0)

    return img_m_uint8
```

## 8. Exécution et visualisation

```python
def crop_face(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    rects = detector(gray, 1)
    if len(rects) == 0:
        return img
    r = rects[0]

    margin = 50
    x1 = max(0, r.left() - margin)
    y1 = max(0, r.top() - margin)
    x2 = min(img.shape[1], r.right() + margin)
    y2 = min(img.shape[0], r.bottom() + margin)
    return img[y1:y2, x1:x2]

path1 = 'image1.jpg'
path2 = 'image2.jpg'

img_a = cv2.imread(path1)
img_b = cv2.imread(path2)

if img_a is not None and img_b is not None:

    img_a = crop_face(img_a)
    img_b = crop_face(img_b)

    #Même taille pour les deux
    img_a = cv2.resize(img_a, (600, 600))
    img_b = cv2.resize(img_b, (600, 600))

    resultat = face_morph(img_a, img_b, 0.5)

    if resultat is not None:

        plt.figure(figsize=(15, 5))

        plt.subplot(1, 3, 1)
        plt.imshow(cv2.cvtColor(img_a, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.title("Image 1")

        plt.subplot(1, 3, 2)
        plt.imshow(cv2.cvtColor(resultat, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.title("Morphing (50%)")

        plt.subplot(1, 3, 3)
        plt.imshow(cv2.cvtColor(img_b, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.title("Image 2")

        plt.tight_layout()
        plt.show()
    else:
        print(" Points du visage non détectés.")
else:
    print(" Fichiers images introuvables.")
```
